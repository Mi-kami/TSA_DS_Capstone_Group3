The data shows a strong right skew. Most prices are below NGN 1,000, with a long tail extending up to NGN 6,000 or more. This is common in commodity price data that includes both low-cost items like Maize and Tomatoes, and more expensive ones like Rice, Beans, and Yam.

The distribution looks somewhat bimodal.
 There are two noticeable peaks—one main peak between NGN 200 and 500, mainly from Maize and Tomatoes, and a smaller peak between NGN 1,000 and 2,000, driven by Rice, Beans, and Yam. If we look at each commodity separately, as planned, these two groups will become clear.

There are also some extreme values.

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='commodity', y='price_ngn', data=pi, color="yellow")
plt.xticks(rotation=45)
plt.title("Price Distribution per Commodity")
plt.show()
```

White Maize and Yellow Maize are the cheapest foods in the dataset, with an average cost of about NGN 150 to 180. Their prices stay pretty much the same over time, which makes them the easiest to predict.
Imported Rice is the most expensive item, usually costing around NGN 1,500, but it can go up as high as NGN 6,000 in some places after 2023. This happens because the price of imported rice depends a lot on the exchange rate — when the naira weakens, imported rice becomes very costly very quickly.
Local Rice is cheaper than imported rice, typically around NGN 1,000, but it follows a similar pattern — when imported rice goes up, local rice also tends to go up.
Yam prices change a lot depending on the season and the state. It can be as low as NGN 300 or as high as NGN 900 in normal times, but it can spike even more after bad harvests or when there is conflict.
White Beans have the most unpredictable prices of all — the difference between the cheapest and the most expensive months is larger than any other commodity. This makes it the hardest to forecast accurately.
Tomatoes are generally the cheapest item, but they can suddenly become very expensive — this is probably because of supply problems in areas affected by conflict, like Borno and Yobe.
Bottom Line
There are clearly two groups of commodities — the cheap ones like Maize and Tomatoes, and the more expensive ones like Rice, Beans, and Yam.

## correlation matric 

```python
numeric_cols = [
    'price_ngn', 'rainfall_mm', 'pms_price_ngn',
    'conflict_score_weighted', 'conflict_fatalities',
    'inflation_rate_pct', 'exchange_rate_ngn_usd',
    'mpr_pct', 'import_volume_tonnes',
    'price_lag1', 'price_lag2', 'price_lag3',
    'other_commodity_avg_price'
]

corr = pi[numeric_cols].corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

The heatmap shows how closely each variable is connected to food prices.

Red means there's a strong positive link, meaning when one goes up, the other tends to go up too.
Blue means there's a strong negative link, meaning when one goes up, the other tends to go down.
Light or white colors mean there's little to no connection

 Key Findings

Lag features (price_lag1, lag2, lag3) are the best predictors (very high connection).
 The price from last month is the best sign of what the price will be this month.
Exchange rate (0.63) and PMS petrol price (0.61) have strong positive links.
When the naira gets weaker or fuel prices go up, food prices also go up.


Inflation (0.58) has a moderate to strong positive link.
 Overall economic problems tend to make food prices go up.
Import volume (-0.43) is the only meaningful negative link.
More imports mean more supply, which can help lower prices.
Conflict (0.14) and Rainfall (0.046) have weak effects nationally.
 Their influence might be bigger at the state level but not when looking at the whole country.


 Bottom Line

Lag features are the strongest predictors
Exchange rate, petrol price, and inflation are the next most important factors.
Import volume helps bring prices down.
Rainfall and conflict don't add much value at the national level.
This gives clear direction on which features should be used in the forecasting model.

## Outlier Detection and Data Coverage

```python
import numpy as np

def detect_outliers(group):
    Q1 = group['price_ngn'].quantile(0.25)
    Q3 = group['price_ngn'].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR
    
    outliers = group[
        (group['price_ngn'] < lower_bound) |
        (group['price_ngn'] > upper_bound)
    ]
    
    return outliers
```

```python
outliers = pi.groupby(['state', 'commodity']).apply(detect_outliers)

print("Total Outlier Rows:", len(outliers))
print(outliers[['date', 'state', 'commodity', 'price_ngn']])
```

```python
outlier_count = outliers.reset_index(drop=True) \
                         .groupby('commodity')['price_ngn'] \
                         .count()

print("\nOutliers per Commodity:")
print(outlier_count.sort_values(ascending=False))
```

The IQR method was used to find extreme price values in each state-commodity group.
A few extreme price spikes were found.

These unusual prices probably show real economic problems like currency value drops, ending of government support, or supply issues, not mistakes in the data.

Because of this, the unusual prices were noted but not taken out so that real market changes could be studied properly.

## What This Section Does

- Spots big price jumps  
- Checks that they are noted but not deleted  
- Points out where data is missing  
- Highlights potential issues with the model

```python
from statsmodels.tsa.stattools import adfuller

def check_stationarity(series, name):
    result = adfuller(series.dropna())
    
    print(f"\n{name}")
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    
    if result[1] < 0.05:
        print("Result: Series is STATIONARY ✅")
    else:
        print("Result: Series is NOT STATIONARY ❌ (Differencing needed)")
```

```python
subset = pi[(pi['commodity'] == 'Imported Rice') & 
            (pi['state'] == 'Kano')]

series = subset.set_index('date')['price_ngn']

check_stationarity(series, "Imported Rice - Kano")
```

```python
series_diff = series.diff()

check_stationarity(series_diff, "Differenced Series")
```
