## **Descriptive Satistics**
The Table above shows a descriptive statistics for the agricultural food supply. This indicates that the datapoints are widely spread out from the mean, showing high variability. Practically, the prices are more inconsistent, and extreme values occur more frequently.

The average amount of rain farmers can expect in a month is 79.204202 mm. The standard deviation is 115.320231 mm which is much larger than the mean. This indicates rainfall is highly inconsistent from month to month. Some months may have very little rain, while others may experience heavy downpours. Such high variability makes it difficult to plan planting and harvesting schedules. Crops may suffer from drought in some months and flooding in others. Also, inconsistent rainfall can reduce yields, affect crop quality, and increase the risk of food insecurity. In other words, the average rainfall seems adequate for agriculture while the extreme variability means farmers face high risk and uncertainty.

The average monthly petrol pump price 205.018257 naira which is the typical cost farmers face when purchasing fuel for machinery, irrigation,transportation etc. The standard deviation (190.648470 naira) which is nearly as large as the mean indicates extreme volatility in petrol prices from month to month.

Such swings could be driven by supply shocks like poor harvests, transport disruptions, inflation or even demand surges. For consumers, this means food affordability is highly unstable. For farmers, it reflects unstable income and market unpredictability.

The lowest monthly rainfall is 0mm and the maximum is 551.2 mm. Months with 0 mm rainfall indicate drought conditions, which severely limit crop growth. Months with over 500 mm rainfall suggest flooding or excessive rain, which can damage crops and infrastructure. The wide range highlights climate variability, making agriculture highly risky without irrigation or water management systems.

The montly petrol pump price is at 44.56 naira minimum and 1283.79 naira maximum. This shows that fuel costs swing dramatically, affecting farm machinery operation, irrigation, and food transport. High petrol prices increase production and distribution costs, which in turn push food prices upward. Consequently, farmers face uncertainty in planning, as energy costs are a major input in the food supply chain.

Overall, extremme instability in rainfall, fuel prices, and food prices are observed. This combination makes agricultural food supply highly vulnerable, requiring strong risk management and adaptation strategies to ensure stability.

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price_ngn',bins=20)
plt.title('Distribution of Monthly average retail price (price_ngn)')
plt.xlabel('Monthly average retail price (price_ngn)')
plt.ylabel('Count')
plt.grid(axis='y',alpha=0.75)
plt.show()
```

The histogram of monthly average retail price (price_ngn) exhibits a strongly positively skewed distribution, with a high concentration of observations at lower price levels and a long right tail extending toward very high prices. This indicates that for most months, retail food prices remained relatively low to moderate, while events of exceptionally high prices occurred less frequently.

The spread of the distribution is wide, with prices ranging from almost zero to over 6,000 naira, reflecting substantial variability in retail food prices over time. Most observations appear clustered below approximately 1,500 naira, suggesting a typical price range for the majority of the period. Beyond this range, the frequency of observations declines sharply, indicating that very high prices are uncommon but economically significant.

The observed pattern suggests the presence of price spikes, likely driven by supply shocks such as poor harvests due to adverse rainfall, increased transportation costs from fuel price hikes, or some economic pressures.

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='rainfall_mm',bins=20)
plt.title('Distribution of Montly Rainfall (rainfall_mm)')
plt.xlabel('Monthly Rainfall (rainfall_mm)')
plt.ylabel('Count')
plt.grid(axis='y',alpha=0.75)
plt.show()
```

The histogram above shows the distribution of monthly rainfall for the agricultural food supply measuresd in millimeters. The distribution is strongly right skewed with a large concentration of observations at very low rainfall levels, particularly between 0 mm and about 50 mm. This indicates that most months within the agricultural food supply period experienced low rainfall. As rainfall amount increases, the frequency of observations declines steadily, forming a long right tail that extends to values above 500 mm.  

The spread of the distribution is wide, ranging from 0 mm to over 500 mm. This reflects substantial variability in monthly rainfall. While low rainfall month dominate, the presence of moderate rainfall values (approximately 100-300 mm) suggests periods of regular rainy season activity. The few extreme high-rainfall observations represent heavy rainfall events, which occur infrequently but may have significant implications for agricultural production, including flooding, soil erosion, or disruptions to food supply logistics.

In summary, the pattern suggests a rainfall regime characterized by many dry (that is 0 mm) or lightly rainy months and occasional intense rainfall episodes. This high variability and positive skewness highlight the importance of effective water management and risk mitigation strategies in agricultural planning and food supply stability.

These high price observations suggest occasional fuel shocks, likely driven by supply disruptions or policy changes, which substantially increase transportation costs within the agricultural supply chain.

Overall, while petrol prices remain relatively stable during most periods, the wide spread and presence of outliers highlight significant price vollatiity that can negatively impact distribution efficiency and cost stability.
