### 2.1 National Food Security Overview

Nigeria is home to over 220 million people, making it the most populous country in Africa, yet it remains one of the most food-insecure nations on the continent. According to the Food and Agriculture Organization of the United Nations (FAO) and the World Food Programme (WFP), approximately 25 million Nigerians faced acute food insecurity in 2023, a figure that has been rising steadily over the preceding decade. The Global Hunger Index (2023) classifies Nigeria's hunger level as serious, with the country ranking 109th out of 125 nations. The National Bureau of Statistics (NBS) reports that food inflation reached an all-time high of 37.92% in July 2024, eroding the purchasing power of millions of households who spend upwards of 60% of their income on food.

Nigeria's agricultural sector, which employs roughly 35% of the working-age population, is characterised by smallholder dominance, limited mechanisation, and heavy reliance on rain-fed production. The country produces significant quantities of staple commodities including rice, maize, yam, beans, and vegetables, yet remains structurally dependent on food imports — particularly for rice and wheat — to bridge domestic supply gaps. The Central Bank of Nigeria (CBN) reported that Nigeria's food import bill exceeded USD 5 billion annually in recent years, making the country highly exposed to global price shocks, exchange rate movements, and foreign exchange scarcity. This structural vulnerability means that disruptions to either domestic production or import supply chains translate rapidly into retail price spikes experienced by ordinary consumers.

### 2.2 The Northeast Nigeria Crisis: Adamawa, Borno, and Yobe

Among Nigeria's 36 states, the three northeastern states of Adamawa, Borno, and Yobe consistently register the highest rates of acute food insecurity in the country. The Cadre Harmonisé — the regional food security classification framework used across West Africa — routinely places large proportions of the population in these states in Crisis (Phase 3) or Emergency (Phase 4) conditions. In 2023, the Cadre Harmonisé estimated that over 4.3 million people across the Northeast faced acute food insecurity during the lean season, with Borno State accounting for the largest share of this caseload.

The primary driver of this crisis is the Boko Haram insurgency, which began in Borno State around 2009 and subsequently expanded into Adamawa and Yobe. Over more than a decade of conflict, the insurgency has devastated the agricultural foundations of the region. Farmers have been killed, displaced, or prevented from accessing their fields due to security threats, resulting in drastic reductions in cultivated land. The Lake Chad Basin — historically one of Nigeria's most productive agricultural zones — has seen a severe contraction in farming activity. The Armed Conflict Location and Event Data Project (ACLED) has documented thousands of conflict events in the tri-state region between 2015 and 2024, including attacks on rural markets, farmlands, and transportation routes.

The humanitarian consequences have been profound. The United Nations Office for the Coordination of Humanitarian Affairs (OCHA) estimated that the conflict had displaced over 2 million people within Nigeria by 2022, with the majority concentrated in Borno State. These internally displaced persons (IDPs) are predominantly farmers who have lost both their land and their livelihoods, creating a population almost entirely dependent on humanitarian food assistance. Market systems across the Northeast have been severely disrupted: supply chains are fragmented, traders face security risks on key transport corridors, and many rural markets have been abandoned or relocated. The combined effect is a food system that is structurally fragile, geographically constrained, and highly sensitive to further shocks.

### 2.3 Price Volatility and Its Drivers

Food price volatility in Northeast Nigeria is driven by an unusually dense concentration of overlapping structural and transitory factors, making the region one of the most challenging food market environments on the continent. Understanding these drivers is essential context for any price forecasting exercise.

Conflict remains the most immediate volatility driver. Market disruptions caused by insecurity reduce supply, increase transport costs, and limit trader activity, all of which cause localised price spikes that are often disconnected from national price trends. The removal of the fuel subsidy by the Tinubu administration in June 2023 constituted a second major structural shock: transport costs surged almost overnight, and since food distribution across Nigeria is overwhelmingly road-dependent, the cost increase was passed directly onto retail food prices. The World Bank estimated that the subsidy removal contributed to an acceleration in headline inflation, which crossed 30% by early 2024 according to NBS data.

Exchange rate depreciation has compounded these pressures. The naira lost more than 70% of its value against the US dollar between 2022 and 2024 following the liberalisation of the foreign exchange market, dramatically increasing the naira-denominated cost of imported rice, which is a key reference commodity in Nigerian food markets. Seasonal harvest cycles introduce a further layer of predictable volatility: prices typically peak during the lean season (May–August) before the main harvest, and fall after the September–November harvest period. However, in conflict-affected areas, this seasonal rhythm is disrupted by insecurity during planting and harvesting periods, creating unpredictable seasonal price dynamics. Rainfall anomalies — either flooding or drought — further compound supply uncertainty, particularly for rain-dependent crops such as yam and tomatoes.

### 2.4 The Case for Early Warning and Forecasting

Given this complex and compounding set of vulnerabilities, early warning systems are not merely useful but operationally critical. Humanitarian organisations such as WFP and NGOs operating in the Northeast must make procurement and pipeline decisions weeks or months in advance. Policymakers at the federal and state levels require lead time to activate strategic grain reserves, coordinate import licences, or deploy cash transfer programmes before a price crisis renders food inaccessible to the most vulnerable populations. Traders and market actors similarly benefit from price signals that allow them to make stocking and logistical decisions that can dampen rather than exacerbate price spikes.

A three-month forecasting horizon is specifically chosen in this project because it represents the minimum operational planning window for most humanitarian response actors while remaining within a temporal range where price dynamics are meaningfully predictable. Forecasting beyond six months in a conflict environment introduces too much uncertainty due to the non-stationarity of conflict drivers; forecasting only one month ahead provides insufficient lead time for logistical mobilisation.

### 2.5 Data Sources and Relevance

This project draws on five core data sources, each selected for its direct relevance to the food price dynamics described above. The WFP Vulnerability Analysis and Mapping (VAM) retail price dataset provides monthly retail prices for the five target commodities across markets in Adamawa, Borno, and Yobe, and constitutes the primary target variable for all forecasting models. CBN macroeconomic indicators — including exchange rates and money supply data — capture the monetary and trade dynamics that transmit global price shocks into domestic food markets. ACLED conflict event data provides a georeferenced record of security incidents that affect agricultural activity and market access in the Northeast. FAO import data tracks Nigeria's commodity import volumes and values, capturing supply-side disruptions driven by foreign exchange constraints. Finally, CHIRPS (Climate Hazards Group InfraRed Precipitation with Station data) rainfall data provides high-resolution precipitation records relevant to agricultural production cycles for rain-dependent commodities. Together, these sources provide a multi-dimensional picture of the drivers of food price variability in one of the world's most complex food crisis environments.

### 2.6 Connecting Context to This Project

The scale, complexity, and human cost of food insecurity in Northeast Nigeria make the development of robust, data-driven price forecasting tools both a technical and a humanitarian imperative. The Early Warning Food Price Monitor — Northeast Nigeria is designed to bridge the gap between existing price monitoring systems, which report historical trends, and the forward-looking intelligence that actionable humanitarian and policy responses demand. By integrating conflict, climate, macroeconomic, and market data into three complementary forecasting models — SARIMA, XGBoost, and Facebook Prophet — this project aims to provide a three-month price outlook for five critical commodities across three of Nigeria's most vulnerable states, giving decision-makers the lead time needed to act before crisis conditions become irreversible.

### 5.3 Feature Engineering

#### Overview

Feature engineering was conducted in two phases.
The first phase occurred in Notebook 03 during
dataset merging, producing three engineered features
that were added to the master dataset and used by
all downstream models. The second phase occurred
within each modelling notebook, where model-specific
transformations were applied immediately before
training.

---

#### Phase 1 — Master Dataset Features (Notebook 03)

Three features were engineered and saved permanently
into the master dataset.

---

**Feature 1 — Price Lag Features**

Food prices exhibit strong autocorrelation — this
month's price is highly predictable from last month's
price. EDA confirmed this with correlation values of
r = 0.968 for price_lag1, r = 0.953 for price_lag2,
and r = 0.924 for price_lag3 against the target
variable. These are the three strongest predictors
in the entire feature set.

Three lag features were created per commodity-state
combination:

| Feature | Definition |
|---------|-----------|
| price_lag1 | Retail price 1 month prior |
| price_lag2 | Retail price 2 months prior |
| price_lag3 | Retail price 3 months prior |

Implementation: The dataset was sorted by state,
commodity and date before applying pandas shift()
within each commodity-state group. Sorting before
shifting is critical — unsorted data would assign
lags from the wrong time period, introducing
data leakage.

The first 1, 2 and 3 months of each commodity-state
series produce null values representing the warmup
period where insufficient history exists. These nulls
(71 for lag1, 141 for lag2, 211 for lag3) were
resolved in each modelling notebook using forward
fill and backward fill within groups immediately
before model training.

---

**Feature 2 — Harvest Season Indicator**

Nigerian agricultural commodities follow predictable
annual harvest cycles. During harvest months, supply
increases and retail prices typically fall. A binary
harvest_season indicator tells each model whether
a given observation falls in a harvest month for
that commodity.

Harvest month definitions were set according to the
Nigerian agricultural calendar:

| Commodity | Harvest Months |
|-----------|---------------|
| Maize (white) | August, September, October |
| Maize (yellow) | August, September, October |
| Rice (imported) | October, November, December |
| Rice (local) | October, November, December |
| Beans (white) | October, November |
| Yam | August, September, October |
| Tomatoes | November, December, January |

Implementation: A dictionary mapping each commodity
to its harvest months was defined. A row-wise lambda
function checks whether the month of each row falls
within the harvest months for that row's commodity
and assigns 1 (harvest) or 0 (non-harvest).

Result: 727 harvest month rows (25.5%) and 2,119
non-harvest rows (74.5%) — consistent with a cycle
of 2–3 harvest months per commodity across a
12-month year.

Note: EDA revealed that harvest_season showed
weaker-than-expected correlation with retail prices.
Market integration and storage dynamics meant that
retail prices do not always fall immediately during
harvest months. This was documented in EDA Step 8
and the feature was retained as a directional signal
rather than a strong predictor.

---

**Feature 3 — Other Commodity Average Price**

Food commodity prices influence each other through
substitution and complementarity effects. When one
staple becomes expensive, consumers shift demand
toward cheaper alternatives, putting upward pressure
on substitute commodities. Farmers also make
planting decisions based on relative commodity
profitability.

This feature captures aggregate cross-commodity
price pressure in each state and month as a single
summary signal.

Implementation: The mean price across all commodities
was computed for each state-month combination using
groupby mean. This aggregate was then merged back
onto the master dataset on date and state. Each row
receives the average price of all other commodities
in its state and month.

---

#### Phase 2 — Model-Specific Transformations
(Applied in Modelling Notebooks)

Several additional transformations were applied
within individual modelling notebooks rather than
at the master dataset level. This was deliberate —
these transformations are model-specific and applying
them to the master dataset would have coupled
modelling decisions to the shared data layer.

---

**Log transformation of target variable (NB05, NB06, NB07)**

The target variable price_ngn was log-transformed
using log1p (natural log of price + 1) before
model training in all three modelling notebooks.
```python
df['log_price'] = np.log1p(df['price_ngn'])
```

Reason: Food prices are right-skewed with occasional
extreme spikes. Log transformation compresses the
scale, stabilises variance, and converts multiplicative
price growth into additive increments — a more
appropriate representation for regression models.
All predictions were inverse-transformed back to
Naira scale using expm1 before metric calculation
to ensure MAPE and RMSE are interpretable.

---

**Subsidy removal indicator (NB05, NB06, NB07)**

A binary structural break indicator was engineered
to explicitly mark the June 2023 fuel subsidy removal
— the single most consequential event in the dataset.
```python
df['subsidy_removed'] = (df['date'] >= '2023-06-01').astype(int)
```

The value is 0 for all observations before June 2023
and 1 from June 2023 onward. This feature gives each
model a direct signal that price behaviour changed
structurally at this point. In Prophet it was used
as an external regressor. In XGBoost it was included
in the feature set.

Note: A single known changepoint date of June 2023
was initially considered for Prophet's built-in
add_changepoints parameter but was rejected because
Prophet rejects changepoint dates that fall outside
the training window boundary. The external regressor
approach was used instead.

---

**Log transformation of import volume (NB06)**

Import volume (import_volume_tonnes) was
log-transformed before use in XGBoost to correct
its heavily right-skewed distribution identified
in EDA.
```python
df['log_import_volume'] = np.log1p(
    df['import_volume_tonnes'].clip(lower=0)
)
```

The clip(lower=0) guard prevents negative values
from causing log transformation errors.

---

**Categorical encoding of state and commodity (NB06)**

XGBoost requires numerical inputs. State and commodity
were label-encoded using sklearn's LabelEncoder.
```python
df['state_encoded']     = LabelEncoder().fit_transform(df['state'])
df['commodity_encoded'] = LabelEncoder().fit_transform(df['commodity'])
```

Label encoding was chosen over one-hot encoding
to keep the feature space compact given the small
per-combination training sizes of 60–90 observations.

---

#### Feature Exclusions and Rationale

Two features were collected and merged into the
master dataset but explicitly excluded from all
modelling:

| Feature | Correlation | Retained Feature | Reason for Exclusion |
|---------|-------------|-----------------|---------------------|
| conflict_events | r = 0.99 with conflict_score_weighted | conflict_score_weighted | Near-identical information. Retaining both splits feature importance without adding signal |
| mpr_pct | r = 0.88 with pms_price_ngn | pms_price_ngn | High redundancy with fuel price. Retaining both dilutes the fuel price importance signal |

---

#### Final Feature Set Summary

| Feature | Engineering Phase | Type |
|---------|------------------|------|
| price_lag1, price_lag2, price_lag3 | NB03 | Lag features |
| harvest_season | NB03 | Calendar indicator |
| other_commodity_avg_price | NB03 | Cross-commodity signal |
| log_price | NB05/06/07 | Target transformation |
| subsidy_removed | NB05/06/07 | Structural break indicator |

### 5.4 Model Architecture and Implementation

#### Overview

Three forecasting models were built and evaluated:
SARIMA (Notebook 05), XGBoost (Notebook 06), and
Facebook Prophet (Notebook 07). A naive baseline
was established identically in all three notebooks
to ensure all comparisons are made on equal terms.
Each model was evaluated on the same Tier 1
commodity-state combinations covering Adamawa,
Borno and Yobe. Prophet was additionally extended
to Tier 2 states.

---

#### Naive Baseline

The naive baseline predicts that next month's price
equals the last known price before the test period.
It requires no training and makes no assumptions
about structure. It serves as the minimum performance
floor — any model that cannot outperform it has
no practical value.

The same naive baseline logic was implemented
identically in NB05, NB06 and NB07 to guarantee
fair comparison across models.

---

#### Model 1 — SARIMA (Notebook 05)

SARIMA — Seasonal AutoRegressive Integrated Moving
Average — is the classical statistical baseline
for time series forecasting. It models price as
a function of its own past values and past
forecast errors, with an explicit seasonal
component.

SARIMA parameters were determined individually
per commodity-state combination rather than using
a single global parameter set. This was done
because different commodities exhibit different
autocorrelation structures and seasonal patterns.

SARIMAX — the exogenous variable extension —
was also tested across three feature configurations:
macroeconomic features only, conflict features
added, and all features combined. In all three
configurations SARIMAX failed to consistently
outperform the baseline SARIMA, and SARIMA
was retained as the statistical model.

SARIMA was evaluated on Tier 1 only (13
commodity-state combinations) using single-step
forecasting. Multi-horizon evaluation was not
implemented in this notebook — this is a known
limitation documented in the model comparison.

---

#### Model 2 — XGBoost (Notebook 06)

XGBoost — Extreme Gradient Boosting — is a
tree-based ensemble model that learns non-linear
relationships between features and the target.
Unlike SARIMA it can simultaneously incorporate
all engineered features alongside price history.

**Direct multi-horizon forecasting:**
Rather than using recursive forecasting (feeding
predictions back as inputs), three separate
XGBoost models were trained per commodity-state
combination — one each for the 1-month, 2-month
and 3-month ahead horizons. This direct approach
avoids compounding prediction errors that occur
in recursive strategies.

**Feature set used:**
price_lag1, price_lag2, price_lag3,
harvest_season, other_commodity_avg_price,
log_import_volume, conflict_score_weighted,
conflict_fatalities, exchange_rate_ngn_usd,
inflation_rate_pct, pms_price_ngn,
rainfall_mm, subsidy_removed,
state_encoded, commodity_encoded

**Target variable:** log_price (log1p transformed),
inverse-transformed before metric calculation.

XGBoost was evaluated on Tier 1 only
(13 combinations × 3 horizons = 39 models).

---

#### Model 3 — Facebook Prophet (Notebook 07)

Prophet is a decomposable time series model
developed by Meta that models price as the
sum of trend, seasonality and external regressor
components. It is designed to handle structural
breaks and missing data robustly.

**Key configuration decisions:**

| Setting | Choice | Reason |
|---------|--------|--------|
| Seasonality mode | Multiplicative | Food prices scale proportionally — a 20% seasonal effect at ₦500 is different from 20% at ₦2000 |
| Yearly seasonality | Enabled | All commodities show annual cycles confirmed in EDA |
| External regressor | subsidy_removed | Captures June 2023 structural break as a known binary shift |
| Changepoints | Not manually set | Prophet rejected a June 2023 changepoint date falling outside the training window boundary |

**Walk-forward evaluation:**
Prophet implemented walk-forward evaluation at
1, 2 and 3-month horizons — an improvement over
SARIMA's single-step evaluation that enables
direct horizon-by-horizon comparison with XGBoost.

**Tier 1 results:** 13/13 combinations fitted
successfully. Prophet outperformed the naive
baseline on all 13 combinations.

**Tier 2 extension:**
Prophet was extended to all 10 Tier 2 states
(44 commodity-state combinations). Because fixed
date split boundaries produced zero test
observations for some Tier 2 combinations with
shorter coverage, a proportional split function
(get_split_tier2) was implemented that allocates
a fixed percentage of each series to training
and testing regardless of its absolute length.

A credibility filter was applied to Tier 2 results:
combinations with average MAPE above 35% were
classified as directional-only rather than
production-ready. 24 of 44 Tier 2 combinations
passed the credibility filter.



### 5.5 Evaluation Framework

#### Primary Metric — MAPE

Mean Absolute Percentage Error was used as the
primary evaluation metric across all models and
all comparisons.

MAPE = (1/n) × Σ |actual - predicted| / actual × 100

MAPE was chosen because it is scale-independent —
it produces comparable error figures across
commodities with very different price levels.
A 20% MAPE for Yam (higher absolute price) and
a 20% MAPE for Beans (lower absolute price)
represent equivalent forecasting accuracy.
This makes cross-commodity comparison meaningful.

All MAPE calculations were performed on the
original Naira scale after inverse-transforming
log predictions back to prices using expm1.
This is important — computing MAPE on log-scale
predictions would produce artificially deflated
error figures.

---

#### Secondary Metrics — MAE and RMSE

**MAE (Mean Absolute Error)** measures average
absolute prediction error in Naira. It is
directly interpretable — an MAE of ₦200 means
predictions are on average ₦200 from the
actual price.

**RMSE (Root Mean Squared Error)** penalises
large errors more heavily. It is sensitive to
spikes and regime shifts, making it a useful
complement to MAPE for identifying whether
errors are uniformly distributed or concentrated
in high-volatility periods.

---

#### Multi-Horizon Evaluation

Forecasts were evaluated at three separate
horizons:

| Horizon | Definition |
|---------|-----------|
| H1 | 1-month ahead forecast |
| H2 | 2-month ahead forecast |
| H3 | 3-month ahead forecast |

Evaluating across horizons is critical for
assessing whether model accuracy degrades
predictably as the forecast window extends.
Healthy degradation (H1 < H2 < H3 in MAPE)
signals the model is capturing genuine
uncertainty rather than overfitting to the
test period.

SARIMA implemented single-step forecasting only
(H1 equivalent). Its results were reported under
H1_MAPE with H2 and H3 set to null in the model
comparison notebook to avoid misleading
aggregations.

---

#### Baseline Comparison

All model results were compared against the
naive baseline MAPE for each combination.
A model was only considered to add value if
its MAPE was meaningfully lower than the naive
baseline — not marginally.

---

#### Tier 2 Credibility Filter

A MAPE threshold of 35% was applied to Tier 2
Prophet results. Combinations above this threshold
were labelled directional-only, meaning they may
capture the direction of price movement but
cannot be relied upon for precise forecasting.
This distinction was preserved in the results
CSV output for downstream use.

---

#### Model Comparison — Notebook 08

All results were aggregated in Notebook 08 using
three CSV files output from the modelling notebooks:
sarima_results.csv, prophet_results.csv and
xgb_results.csv. The best model per commodity-state
combination was determined by lowest average MAPE
across available horizons. Results were saved as
model_comparison_final.csv and
best_model_per_combination.csv.


### 5.6 Train, Validation and Test Split Design

#### Core Principle

A temporal split was used throughout the project.
Random splitting was explicitly rejected for all
models. In time series forecasting, random splitting
breaks the temporal order of observations, allowing
the model to train on future data and evaluate on
past data — producing optimistic but invalid results.
Every split in this project respects chronological
order: the model never sees any data point that
occurs after the data it is being evaluated on.

---

#### Tier 1 Split — Fixed Date Boundaries

For Tier 1 states (Adamawa, Borno, Yobe) fixed date
boundaries were applied uniformly across all
commodity-state combinations:

| Split | Period | Months |
|-------|--------|--------|
| Training | January 2017 – December 2022 | 72 months |
| Validation | January 2023 – May 2023 | 5 months |
| Test | June 2023 – December 2024 | 19 months |

The training set covers the full pre-subsidy
history. The validation set covers the transition
period before the subsidy removal. The test set
begins at June 2023 — deliberately coinciding with
the fuel subsidy removal — to ensure evaluation
captures the most challenging and policy-relevant
period in the dataset.

The validation set was used for SARIMAX feature
selection in Notebook 05. In Prophet and XGBoost,
walk-forward evaluation was performed on the test
set directly.

---

#### Tier 2 Split — Proportional Boundaries

Tier 2 states have shorter and less uniform data
coverage than Tier 1. Applying the same fixed
date boundaries to Tier 2 combinations produced
zero test observations for multiple combinations
where data ended before June 2023.

A proportional split function (get_split_tier2)
was implemented to resolve this:

- Training: first 70% of each series chronologically
- Validation: next 15%
- Test: final 15%

This guarantees every Tier 2 combination has a
non-empty test set regardless of its absolute
length. The make_future_dataframe call in Prophet
was explicitly extended to cover the full
proportional test period rather than using the
default 30-day horizon.

---

#### Why the Test Period Starts at June 2023

The test set begins at the subsidy removal date
by design. This was a deliberate methodological
decision, not an arbitrary date choice. The period
June 2023 to December 2024 is the hardest
forecasting period in the dataset — prices
accelerated sharply, the naira devalued, and
fuel costs spiked simultaneously. A model that
performs well on this period demonstrates genuine
robustness to structural shocks, not just
