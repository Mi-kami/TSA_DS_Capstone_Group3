# **SECTION 2 - EXECUTIVE SUMMARY**

## Executive Summary

Food prices in Nigeria are volatile, unpredictable, and deeply consequential.
For millions of Nigerian households, a sudden spike in the price of Rice, Maize,
or Tomatoes is not an inconvenience — it is a crisis. For policymakers and
agricultural stakeholders, the inability to anticipate these price movements
means interventions always come too late.

This project builds a machine learning-powered food price forecasting system
that predicts the monthly retail prices of five key commodities — Rice, Maize,
Beans, Yam and Tomatoes — across 13 Nigerian states, three months in advance.

The system is trained on 8 years of historical data spanning 2017 to 2024,
integrating eight carefully curated datasets covering food prices, inflation,
exchange rates, monetary policy, rainfall patterns, fuel prices, import volumes
and conflict events. By combining classical time series models with modern
machine learning approaches, the system captures both the seasonal rhythms of
Nigerian agriculture and the macroeconomic shocks that disrupt them.

The goal is not just accurate prediction — it is actionable intelligence.
A farmer deciding what to plant, a trader planning their inventory, a government
agency designing a food subsidy intervention — all of them benefit from knowing
where prices are headed three months from now. This project puts that power
in their hands.

# **SECTION 3 - PROBLEM STATEMENT**

## 3.1 The Reality of Food Price Volatility in Nigeria

Nigeria is a nation of over 200 million people, the majority of whom spend
a significant proportion of their income on food. Unlike wealthier economies
where food represents a small fraction of household expenditure, food costs
in Nigeria can consume up to 60% of a low-income household's budget. In this
context, food price volatility is not merely an economic indicator — it is a
matter of survival.

What makes Nigeria's food price problem particularly severe is its asymmetry.
Prices rise rapidly in response to shocks — a poor harvest, a naira devaluation,
a fuel subsidy removal, a surge in conflict in farming communities — but they
rarely return to their previous levels. The June 2023 removal of the petrol
subsidy, for example, triggered an immediate cascade of price increases across
all food categories that persist to this day. For the average Nigerian household,
each shock permanently resets the cost of living to a higher level.



## 3.2 Who Bears the Burden

-  **Households** bear the heaviest cost. When the price of Rice, Maize or Tomatoes
spikes beyond what a family can afford, the consequences are immediate — reduced
meal frequency, substitution to lower nutrition alternatives, and in the worst
cases, hunger. Children are disproportionately affected, with food price shocks
directly linked to rising rates of malnutrition in vulnerable states.

- **Traders and market actors** face a different but equally serious challenge.
Without visibility into future price movements, traders cannot plan their
inventory, cannot negotiate fair contracts, and cannot protect their margins.
A trader who stocks up on Tomatoes at harvest time expecting stable prices
can be wiped out by an unexpected supply shock weeks later. Price unpredictability
makes rational business planning nearly impossible.

- **Policymakers and government agencies** are perpetually reactive. Food price
crises trigger protests, social unrest and political instability. Yet without
early warning systems, government interventions — import waivers, subsidy
programmes, emergency food distributions — are always designed in response
to a crisis that has already begun rather than one that can still be prevented.



## 3.3 The Downstream Consequences

The consequences of unchecked food price volatility extend far beyond hunger:

- **Food insecurity** affects an estimated 25 million Nigerians and worsens
  with every major price shock
- **Poverty entrenchment** as households deplete savings and assets to cover
  basic food costs
- **Rising crime rates** as economic desperation drives individuals toward
  illegal means of meeting basic needs
- **Agricultural disinvestment** as farmers, unable to predict what their
  produce will be worth at harvest, make suboptimal planting decisions that
  perpetuate supply instability



## 3.4 Why Existing Solutions Are Insufficient

Current approaches to monitoring food prices in Nigeria are largely retrospective.
The National Bureau of Statistics publishes monthly price data — but by the time
data is collected, processed and published, the price movement it describes has
already happened. There is no forward-looking intelligence.

Academic research on Nigerian food prices has focused primarily on identifying
drivers of price changes rather than building operational forecasting tools.
Studies have analysed the impact of exchange rates, rainfall and conflict on
food prices in isolation — but no comprehensive multi-feature forecasting system
exists that integrates all these drivers simultaneously at the state level.

International tools like the FAO Food Price Index operate at a global or
national level and cannot capture the significant regional variation in food
prices across Nigerian states. A model trained on national averages is of
limited use to a trader in Kano or a policymaker in Borno.

This project addresses all three gaps — it is forward-looking, it integrates
multiple drivers simultaneously, and it operates at the state level where
food price decisions are actually made.

# **SECTION 4- LITERATURE REVIEW**

## 4.1 Global Research on Food Price Forecasting

Food price forecasting has been an active area of research globally,
driven by the recognition that price volatility threatens food security,
macroeconomic stability and political order. Early approaches relied
heavily on classical econometric models. Autoregressive Integrated Moving
Average (ARIMA) models, introduced by Box and Jenkins in 1970, became the
standard baseline for time series forecasting and have been widely applied
to agricultural commodity prices. Studies such as Isenah and Olubusoye (2014)
demonstrated ARIMA's utility in modelling Nigerian inflation series,
establishing a foundation for time series approaches in the Nigerian context.

Seasonal ARIMA (SARIMA) extended this foundation by explicitly capturing
recurring seasonal patterns — particularly relevant for agricultural
commodities where harvest cycles create predictable annual price rhythms.
Research across Sub-Saharan Africa confirmed that food prices exhibit strong
seasonal components that purely trend-based models fail to capture adequately.


## 4.2 The Rise of Machine Learning in Food Price Forecasting

The limitations of classical statistical models — particularly their
inability to incorporate multiple external drivers simultaneously —
drove researchers toward machine learning approaches. Gradient boosting
algorithms, particularly XGBoost, emerged as powerful tools for commodity
price prediction due to their ability to handle complex non-linear
relationships between economic indicators and prices.

Cao et al. (2020) demonstrated that XGBoost significantly outperformed
ARIMA in multi-step ahead food price forecasting when macroeconomic
features were included as inputs. Similarly, research on East African
food markets showed that models incorporating rainfall, fuel prices and
conflict indicators consistently outperformed models using price history
alone — a finding directly relevant to the Nigerian context.

Facebook Prophet, introduced by Taylor and Letham (2018), offered a
different paradigm — a decomposable time series model designed for
analysts rather than statisticians. Its ability to handle missing data,
outliers and multiple seasonality patterns made it particularly attractive
for developing country contexts where data quality is inconsistent.


## 4.3 Food Price Forecasting in Nigeria and West Africa

Research specifically focused on Nigerian food prices has grown
significantly since the 2008 global food crisis. Nweze and Edeh (2016)
examined the relationship between exchange rate movements and food prices
in Nigeria, finding strong transmission effects — particularly for imported
commodities like rice. Adekunle et al. (2020) demonstrated that petroleum
price shocks significantly amplify food price volatility through their
effect on transportation and production costs.

Studies on the impact of conflict on food prices in Nigeria's Northeast —
particularly in Borno and Yobe states affected by insurgency — confirmed
that security disruptions create persistent price premiums in affected
markets that can last months after active conflict subsides. Olomola (2019)
highlighted the role of rainfall variability in driving food price
instability across Nigeria's agricultural belt, noting that states with
rain-fed farming systems show significantly higher price volatility during
drought years.

Despite this body of research, a critical gap persists — existing studies
analyse individual drivers of food price changes but do not integrate them
into operational forecasting systems. The research explains what causes
price movements but does not predict them.


## 4.4 Identified Gaps in Existing Literature

A systematic review of existing food price forecasting research in Nigeria
and West Africa reveals four critical gaps that this project directly addresses:

**Gap 1 — Absence of Multi-Feature Integration:**
Existing models typically examine one or two drivers of food price changes.
No study has built a comprehensive forecasting model that simultaneously
integrates macroeconomic indicators, agricultural inputs, supply chain
costs, conflict data and import volumes at the state level.

**Gap 2 — National Level Analysis Masking Regional Variation:**
The majority of Nigerian food price research operates at the national
level using aggregated data. This masks significant regional variation —
the price of Maize in Zamfara behaves very differently from the price of
Maize in Lagos due to different supply chains, security situations and
market structures. State-level forecasting models are virtually absent
from the literature.

**Gap 3 — Retrospective Rather Than Predictive Focus:**
Existing research is largely explanatory — it tells us why prices moved
after the fact. Operational forecasting tools that provide actionable
forward-looking predictions three months ahead are not available for
Nigerian food markets.

**Gap 4 — Limited Multi-Model Comparison:**
Few studies compare classical time series models, gradient boosting
algorithms and modern decomposable forecasting frameworks on the same
Nigerian food price dataset. Without such comparison it is impossible
to identify which approach best captures the unique dynamics of Nigerian
food markets.


## 4.5 How This Project Addresses the Gaps

| Gap | Our Response |
|-----|-------------|
| No multi-feature integration | 8 datasets integrated covering macro, agricultural, supply chain and security dimensions |
| National level only | State-level forecasting for 13 Nigerian states |
| Retrospective focus | 3-month ahead operational forecasting |
| Limited model comparison | ARIMA, SARIMA, XGBoost and Prophet compared on identical data |

This project does not simply replicate existing approaches — it builds
on their foundations while directly addressing the limitations that have
prevented food price forecasting from becoming an operational tool in
the Nigerian context.


# **SECTION 5 - PROJECT OBJECTIVES**

## Project Objectives

This project is guided by the following clear and measurable objectives:

**Objective 1 — Build a State-Level Food Price Forecasting System**
To develop a machine learning system capable of forecasting monthly
retail prices for five key food commodities — Rice, Maize, Beans, Yam
and Tomatoes — across 13 Nigerian states, three months ahead with
demonstrable accuracy.

**Objective 2 — Achieve Measurable Forecasting Accuracy**
To build models that achieve a Mean Absolute Error (MAE) below 10%
of the average commodity price and a Root Mean Squared Error (RMSE)
that outperforms a naive baseline forecast — where the baseline simply
predicts next month's price equals this month's price.

**Objective 3 — Integrate Multiple Economic Drivers**
To go beyond price history alone by incorporating eight distinct
data sources — macroeconomic indicators, agricultural inputs, supply
chain costs, conflict data and import volumes — into a unified
forecasting framework that captures the full complexity of Nigerian
food price dynamics.

**Objective 4 — Compare Multiple Modelling Approaches**
To evaluate and compare four modelling approaches — ARIMA, SARIMA,
XGBoost and Facebook Prophet — on identical data to identify which
approach best captures Nigerian food price behaviour and provide
evidence-based justification for the final model selection.

**Objective 5 — Generate Actionable Intelligence**
To produce forecasts that are practically useful — not just
academically accurate. The system should provide insights that
help households plan food budgets, traders manage inventory,
farmers make planting decisions, and policymakers design timely
interventions before price crises escalate.

**Objective 6 — Contribute to Food Security**
To demonstrate that data-driven forecasting can serve as an early
warning system for food price shocks — enabling proactive responses
that reduce food insecurity, moderate the burden on low-income
households and contribute to broader economic stability in Nigeria.

**Objective 7 — Establish a Replicable Methodology**
To document a clean, reproducible data science pipeline that can
be extended to additional states, commodities or time periods —
laying the groundwork for a scalable national food price
intelligence system.


# **SECTION 6 - SCOPE AND LIMITATIONS**

## 6.1 Scope

**Geographic Coverage — 13 States**
This project covers 13 Nigerian states: Abia, Adamawa, Borno, Gombe,
Jigawa, Kaduna, Kano, Katsina, Kebbi, Lagos, Oyo, Yobe and
Zamfara. These states were selected based on data availability — they
represent the complete coverage of the WFP food price monitoring
programme in Nigeria, which is the only publicly available source of
consistent historical retail food price data at the state level.
The 13 states span four geopolitical zones and include both major
urban markets and rural agricultural communities, providing
meaningful geographic diversity.

**Commodity Coverage — 7 Commodities**
The seven commodities — Maize (white), Maize (yellow), Rice (imported), Rice (local),
Beans (white), Yam and Tomatoes — were
selected because they represent the most widely consumed staple foods
in Nigeria and have sufficient data coverage across all 13 states
throughout the study period. These commodities collectively reflect
both imported and domestically produced food, both perishable and
non-perishable items, and both grain and vegetable categories —
providing a representative cross-section of the Nigerian food basket.

**Time Period — 2017 to 2024**
The study period spans eight years — 96 monthly observations per
state per commodity. This range was selected because:
- 2017 represents the earliest year with consistent data availability
  across all eight datasets
- 2024 is the most recent year with complete data across all sources
- Eight years provides sufficient historical depth for time series
  models to learn seasonal patterns, economic cycles and structural
  breaks including the COVID-19 period and the 2023 subsidy removal


## 6.2 Limitations

**Limitation 1 — Geographic Coverage Gap**
Two geopolitical zones — North Central and South South — are not
covered due to absence of WFP food price data. States including
Abuja, Rivers, Delta, Cross River and Edo are excluded. This means
the model cannot make predictions for these states and findings
should not be generalised to them without further data collection.

**Limitation 2 — Rainfall Data Gaps**
CBN rainfall data for Abia and Jigawa states is only available
from 2023. Missing values for earlier periods were handled through
forward filling. This introduces some uncertainty in rainfall-related
predictions for these two states specifically.

**Limitation 3 — Annual Import Volume Data**
FAO import volume data is published annually not monthly. Monthly
estimates were derived by distributing annual values evenly across
12 months. This assumption of uniform monthly distribution may not
reflect actual seasonal import patterns.

**Limitation 4 — PMS Price Gap Period**
NBS petrol price data has a gap from June 2019 to August 2021.
This gap was filled using scaled Brent crude oil prices. While
methodologically sound the scaled values are estimates rather than
observed Nigerian prices.

**Limitation 5 — Model Scope**
This project forecasts retail market prices — the prices consumers
pay in markets. It does not forecast farm gate prices, wholesale
prices or processed food prices. The findings are most directly
applicable to retail market actors and household food budget planning.

**Limitation 6 — Sokoto Excluded Due to Wholesale-Only Coverage**
WFP food price records for Sokoto state only contain wholesale
prices. Since this project focuses on retail prices paid by
households, Sokoto could not be included without compromising
price type consistency across the dataset. The final model
covers 13 states instead of the initially planned 14.

## 6.3 What This Project Does Not Do

| Out of Scope | Reason |
|-------------|--------|
| Forecast prices for all 36 states | No consistent food price data exists for 22 states |
| Daily or weekly price forecasting | Data granularity is monthly across all sources |
| Forecast processed food prices | WFP monitors raw commodity prices only |
| Real-time forecasting system | Requires live data pipeline beyond project scope |
| Causal analysis of price drivers | Focus is predictive accuracy not causal inference |

## 7.1 Justification for Each Data Source

- **WFP Food Prices — Primary Dataset**
The WFP HDX food price monitoring programme is the only publicly
available source of consistent, verified, state-level retail food
price data in Nigeria. It is collected by trained field enumerators
across multiple markets per state and is the global standard for
food price monitoring in developing countries. No alternative source
provides comparable coverage, consistency or credibility.

- **CBN Inflation Rate**
The Central Bank of Nigeria Statistical Bulletin is the authoritative
source for Nigerian macroeconomic data. Food inflation directly
reflects the national price environment and is a leading indicator
of retail price movements. Using the food-specific inflation index
rather than headline inflation ensures the feature is directly
relevant to food price dynamics.

- **CBN Exchange Rate**
Nigeria imports significant quantities of Rice, Wheat and other
food commodities. When the naira depreciates against the dollar
import costs rise and this is transmitted to retail prices —
particularly for imported commodities. The 2023 naira devaluation
and its catastrophic effect on food prices makes this one of the
most critical features in the model.

- **CBN Monetary Policy Rate**
The MPR influences borrowing costs for agricultural producers and
food importers. High interest rates reduce investment in food
production and increase the financing cost of food imports —
both of which constrain supply and push prices higher. It also
reflects the broader monetary policy stance that shapes the
inflationary environment.

- **CBN Rainfall Data**
Nigeria's agricultural sector is predominantly rain-fed. Rainfall
is the single most direct determinant of domestic food supply.
Poor rainfall in key growing months leads to harvest failures
that reduce supply and drive prices upward. State-level rainfall
data allows the model to capture regional agricultural conditions
rather than relying on national averages.

- **NBS PMS Petrol Price**
Food cannot reach markets without transportation and transportation
in Nigeria runs on petrol. The June 2023 fuel subsidy removal
demonstrated with brutal clarity how directly petrol prices
transmit into food prices. This feature captures the cost of
moving food from farms to markets — a transmission mechanism
unique to the Nigerian context.

- **FAO Import Volume**
Import volume measures the quantity of each commodity entering
Nigeria from global markets. High import volumes indicate
sufficient supply which moderates price increases. Low import
volumes signal potential supply gaps. This is particularly
relevant for Rice where imported and locally produced varieties
compete directly in Nigerian markets.

- **ACLED Conflict Data**
Nigeria's food belt — spanning the North West, North East and
Middle Belt — has experienced persistent conflict including
bandit attacks, farmer-herder clashes and insurgency. These
events disrupt planting, harvesting and transportation of food.
ACLED is the globally recognised standard for conflict event
data in Africa and provides weekly georeferenced conflict
records that we aggregated to monthly state-level conflict
intensity scores.


# **SECTION 8 - METHODOLOGY OVERVIEW**

## 8.1 The Data Science Pipeline

This project follows a structured end-to-end data science pipeline
consisting of six sequential phases. Each phase builds on the output
of the previous one and produces a saved artifact that feeds into
the next.
```
┌─────────────────────────────────────────────────────────────┐
│                    PROJECT PIPELINE                          │
├──────────────┬──────────────────────────────────────────────┤
│   PHASE 1    │  Data Collection                             │
│              │  Sourcing 8 datasets from 6 institutions     │
├──────────────┼──────────────────────────────────────────────┤
│   PHASE 2    │  Data Cleaning                               │
│              │  Standardising, filtering, reshaping         │
├──────────────┼──────────────────────────────────────────────┤
│   PHASE 3    │  Merging                                      │
│              │  Joining all datasets into master dataframe  │
├──────────────┼──────────────────────────────────────────────┤
│   PHASE 4    │  Feature Engineering                         │
│              │  Price lags, harvest indicator, commodity    │
│              │  cross-prices                                │
├──────────────┼──────────────────────────────────────────────┤
│   PHASE 5    │  Modelling                                    │
│              │  SARIMA → SARIMAX → XGBoost → Prophet          │
├──────────────┼──────────────────────────────────────────────┤
│   PHASE 6    │  Evaluation and Comparison                   │
│              │  RMSE, MAE, feature importance, deployment   │
└──────────────┴──────────────────────────────────────────────┘
```


## 8.2 Phase Descriptions

- **Phase 1 — Data Collection**
Eight datasets were sourced from WFP, CBN, NBS, FAOSTAT and ACLED.
This phase involved significant manual effort — including compiling
37 individual NBS monthly petrol price reports into a single dataset
and navigating complex multi-sheet CBN Statistical Bulletin files.
All raw data was stored in the project Github under the
/data/raw directory.

- **Phase 2 — Data Cleaning**
Each dataset was cleaned independently in a dedicated notebook
section. Cleaning operations included reshaping wide format tables
to long format, standardising date columns to monthly period format,
handling missing values, filtering to the study scope of 14 states
and 2017 to 2024, and aggregating granular records to monthly
averages. All cleaned datasets were saved as CSV files to
/data/processed.

- **Phase 3 — Merging**
All eight cleaned datasets were merged into a single master
dataframe using date and state as the primary merge keys.
Datasets with national coverage — inflation, exchange rate
and MPR — were broadcast across all 13 states. Import volume
was merged on date and commodity. Remaining missing values
were handled through forward filling and zero imputation
where appropriate.

- **Phase 4 — Feature Engineering**
Three categories of features were engineered from the master
dataset. Price lag features capture autocorrelation in price
series — the tendency of this month's price to be predicted
by last month's price. A harvest season indicator captures
the annual supply cycle for each commodity based on Nigerian
agricultural calendars. A related commodity average price
feature captures spillover effects between food categories
within the same state and month.

- **Phase 5 — Modelling**
Four modelling approaches are implemented and evaluated.
SARIMA provides a classical univariate baseline. SARIMAX
extends this to capture seasonal patterns and other patterns with exogenous features. XGBoost leverages
the full feature set as a gradient boosting regressor.
Facebook Prophet provides a decomposable model with
interpretable trend and seasonality components. All models
are trained on data from 2017 to 2022 and evaluated on
2023 to 2024 — a true out-of-sample test.

- **Phase 6 — Evaluation and Comparison**
Models are evaluated using RMSE and MAE on the held-out
test period. Results are compared across all four approaches
and across all 13 states and 5 commodities. Feature
importance analysis identifies which economic drivers
contribute most to forecast accuracy. The best performing
model is selected for deployment.


### 8.3 Reproducibility

All code is version controlled on GitHub. Every notebook
is self-contained — reading inputs from saved CSV files
and writing outputs back to the same directory structure.
Any researcher with access to the repository and the
source datasets can reproduce all results by running
the notebooks in sequence from 01 to 08.

# **SECTION 9 - SUCCESS CRITERIA**

## 9.1 What a Good Model Looks Like

A good forecasting model for this project must satisfy three
criteria simultaneously. First, it must generalise well —
performing consistently on both training data and unseen
test data. A model that fits training data perfectly but
fails on test data is overfit and has no real world value.
Second, it must outperform the baseline meaningfully — not
marginally. Third, it must be interpretable enough that its
predictions can be explained to non-technical stakeholders
like policymakers and traders.

## 9.2 Quantitative Success Metrics

**Primary Metric — Mean Absolute Percentage Error (MAPE)**
MAPE measures the average percentage difference between
predicted and actual prices. It is our primary metric
because it is scale-independent — a MAPE of 20% means
the model's predictions are on average 20% away from
the actual price, regardless of whether the commodity
costs ₦500 or ₦5,000. This makes it directly comparable
across all seven commodity variants and all 13 states
without any adjustment.

Target thresholds for model evaluation:

| MAPE Range | Classification | Operational Use |
|-----------|---------------|-----------------|
| Below 20% | Excellent | Full deployment recommended |
| 20% – 35% | Production-ready | Deployment with monitoring |
| 35% – 100% | Directional only | Signal use — not magnitude |
| Above 100% | Unreliable | Not recommended for use |

A combination is classified as **production-ready** if
it achieves an average MAPE of 35% or below across all
evaluated forecast horizons. This threshold reflects the
reality of forecasting in volatile post-subsidy Nigerian
food markets — where a 35% error on a ₦4,000 commodity
still provides actionable early warning signal for
policymakers and traders.

**Secondary Metric — Mean Absolute Error (MAE)**
MAE measures the average absolute difference between
predicted and actual prices in Nigerian Naira. While
MAPE is used for cross-commodity comparison, MAE is
reported alongside it for interpretability — an MAE of
₦300 on a Rice prediction is directly meaningful to a
trader or food security analyst in a way that a
percentage cannot fully convey.

**Secondary Metric — Root Mean Squared Error (RMSE)**
RMSE penalises large errors more heavily than small ones
by squaring the differences before averaging. It is
particularly important for food price forecasting because
large prediction errors during price spikes — such as the
post-subsidy surge of June 2023 — are operationally more
damaging than small errors during stable periods. A model
with low average MAPE but high RMSE is producing occasional
very large errors that would undermine trust in real
deployment.

**Horizon-Specific Evaluation**
All models are evaluated at three forecast horizons:

| Horizon | Label | Use Case |
|---------|-------|----------|
| H1 | 1-Month Ahead | Immediate market intervention |
| H2 | 2-Month Ahead | Supply chain planning |
| H3 | 3-Month Ahead | Policy and procurement decisions |

Performance is expected to degrade at longer horizons —
a healthy model shows gradual MAPE increase from H1 to H3.
Flat or inverted horizon decay is flagged as a diagnostic
warning sign in each modelling notebook.

**Baseline Requirement**
Every model must outperform the naive baseline — which
simply carries the last known price forward — to justify
its added complexity. A model that cannot beat this simple
benchmark is not adding value regardless of how
sophisticated its algorithm is.

## 9.3 Qualitative Success Criteria

Beyond numbers, we consider the project successful if:

- The model captures **seasonal price patterns** —
  correctly anticipating harvest season price drops
  and lean season price rises
- The model responds appropriately to **economic shocks** —
  reflecting the impact of exchange rate movements,
  fuel price changes and conflict escalation on
  food prices
- The model performs consistently across **multiple
  states and commodities** — not just performing well
  on easy cases while failing on volatile ones
- The forecasting pipeline is **fully reproducible** —
  any team member can run the notebooks and obtain
  identical results
- The findings are **clearly communicated** — through
  visualisations and documentation that a non-technical
  audience can understand and act on

## 9.4 Model Selection Criteria

| Criterion | Weight | Description |
|-----------|--------|-------------|
| Test MAE | High | Primary accuracy measure |
| Test RMSE | High | Penalises large errors |
| Generalisation gap | Medium | Difference between train and test performance |
| Interpretability | Medium | Can results be explained to stakeholders |
| Consistency | Medium | Performance across states and commodities |
| Computational efficiency | Low | Time to train and generate forecasts |

The model that scores best across these criteria —
not just the one with the lowest error — will be
selected as the final model for deployment.
